```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/Students Performance.csv")
df
```

```python
df.info()
```

```python
df.describe(include = 'all')
```

```python
df.groupby(['race/ethnicity']).value_counts()
```

```python
df['test preparation course'].value_counts()
```

```python
df = df.replace({'test preparation course' : {'none' : 0 ,'completed' : 1 }})
df
```

```python
df = df.replace({'gender' : {'female' : 0 ,'male' : 1 }})
df
```

```python
df.columns
```

```python
X = df[['gender',  'math score', 'reading score',
       'writing score']]
X
```

```python
y = df['test preparation course']
y
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X = sc.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.3 , random_state = 456)
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train , y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix , classification_report
```

```python
confusion_matrix(y_test , y_pred)
```

```python
print(classification_report(y_test , y_pred))
```
